```python

import pandas as pd
```

List of tickets sold by A, price at least 30 and less than 40, ordered by type

```python
#1
dfA = pd.read_csv("movie-ticket-seller-A.csv")
dfB = pd.read_csv("movie-ticket-seller-B.csv")

dfA.query("ticket_price>=30 and ticket_price<40").sort_values(by="ticket_type")
```

List of 7 most expensive tickes paid with Card

```python
#2
dfA.query("payment_method=='Credit Card'").sort_values(by="ticket_price",ascending=False).head(7)
```

List of tickets sold by both sellers on seat B2 and sorted by price (only 3 columns)

```python
#3
dfA3=dfA[["movie_title","seat_number","ticket_price"]].query("seat_number=='B2'")
dfB3=dfB[["movie_title","seat_number","ticket_price"]].query("seat_number=='B2'")
dfFull=pd.concat([dfA3,dfB3],ignore_index=True).sort_values(by="ticket_price")
dfFull
```

```python
#4
import matplotlib.pyplot as plt
dfAb=pd.concat([dfA,dfB],ignore_index=True)
dfAb.groupby("ticket_type")["ticket_price"].sum().plot(kind="bar",xlabel="Ticket type",title="Total ticket value for different types of tickets")
```

```python
#5
dfA["seller_type"]="sellerA"
dfB["seller_type"]="sellerB"
dfAb=pd.concat([dfA,dfB],ignore_index=True)
dfAb.groupby(["seller_type"])["ticket_price"].sum().plot(kind="pie",ylabel="",title="Chart of total value of tickets sold by A and B respectively",autopct='%1.0f%%')
```
